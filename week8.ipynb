```python
df['date'] = pd.to_datetime(df['date'], format='ISO8601')
```

# This cell looks at the number of total cases  over time in the country of Kenya,Zimbabwe,India and the United states
# Zimbabwe  had the least number of cases and USA had the most cases 

```python

selected_countries = [country.strip().title() for country in ['United States', 'Kenya', 'Zimbabwe ', 'India']]

df['location'] = df['location'].str.title()

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df_filtered = df[df['location'].isin(selected_countries)]

if df_filtered.empty:
    print("Filtered DataFrame is empty. Check your selected_countries or data.")
else:
    
    df_pivot = df_filtered.pivot(index='date', columns='location', values='total_cases')

   
    if df_pivot.empty:
        print("Pivoted DataFrame is empty. Check your data or filtering logic.")
    else:
        
        df_pivot.plot(figsize=(12, 6))
        plt.title("Total COVID Cases Over Time")
        plt.xlabel("Date")
        plt.ylabel("Total Cases")
        plt.legend(title="Country")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

# This cell shows the daily new cases for Kenya,Zimbabwe,India and the USA
# India and the USA has the highest number of daily cases while kenya and zimbabwe had the least

```python

selected_countries = [country.strip().title() for country in ['Kenya', 'Zimbabwe ', 'India', 'United States']]

df['location'] = df['location'].str.title()

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df_filtered = df[df['location'].isin(selected_countries)]

if df_filtered.empty:
    print("Filtered DataFrame is empty. Check your selected_countries or data.")
else:
    df_pivot = df_filtered.pivot(index='date', columns='location', values='new_cases')

    if df_pivot.empty:
        print("Pivoted DataFrame is empty. Check your data or filtering logic.")
    else:
        df_pivot.plot(figsize=(12, 6))
        plt.title("Daily New COVID Cases Over Time")
        plt.xlabel("Date")
        plt.ylabel("New Cases")
        plt.legend(title="Country")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

# This cell shows covid death rate over time in Kenya,Zimbabwe,India and United States

```python

selected_countries = [country.strip().title() for country in ['Kenya', 'Zimbabwe ', 'India', 'United States']]

df['location'] = df['location'].str.title()

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df['death_rate'] = df['total_deaths'] / df['total_cases']
df['death_rate'] = df['death_rate'].fillna(0)  


df_filtered = df[df['location'].isin(selected_countries)]

if df_filtered.empty:
    print("Filtered DataFrame is empty. Check your selected_countries or data.")
else:
    df_pivot = df_filtered.pivot(index='date', columns='location', values='death_rate')

    if df_pivot.empty:
        print("Pivoted DataFrame is empty. Check your data or filtering logic.")
    else:
        df_pivot.plot(figsize=(12, 6))
        plt.title("COVID Death Rate Over Time")
        plt.xlabel("Date")
        plt.ylabel("Death Rate (Total Deaths / Total Cases)")
        plt.legend(title="Country")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

# This cell shows the total cases over time and the total deaths over time Algeria ,Bahamas ,Zimbabwe and Belgium

```python

selected_countries = [country.strip().title() for country in ['Algeria ', 'Bahamas ', 'Zimbabwe ', 'Belgium ']]

df['location'] = df['location'].str.title()

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df_filtered = df[df['location'].isin(selected_countries)]

if df_filtered.empty:
    print("Filtered DataFrame is empty. Check your selected_countries or data.")
else:
    df_cases_pivot = df_filtered.pivot(index='date', columns='location', values='total_cases')
   
    df_deaths_pivot = df_filtered.pivot(index='date', columns='location', values='total_deaths')

    if df_cases_pivot.empty or df_deaths_pivot.empty:
        print("One or both pivoted DataFrames are empty. Check your data or filtering logic.")
    else:
       
        df_cases_pivot.plot(figsize=(12, 6))
        plt.title("Total COVID Cases Over Time")
        plt.xlabel("Date")
        plt.ylabel("Total Cases")
        plt.legend(title="Country")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        df_deaths_pivot.plot(figsize=(12, 6))
        plt.title("Total COVID Deaths Over Time")
        plt.xlabel("Date")
        plt.ylabel("Total Deaths")
        plt.legend(title="Country")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

# This cell show the total number of deaths per case for 10 countries with the most cases
# Yemen had the highest deaths per case and india had the lowest

```python
df['death_rate'] = df['total_deaths'].div(df['total_cases']).fillna(0)

death_rate_per_country = df.groupby('location')['death_rate'].mean().reset_index()

death_rate_per_country = death_rate_per_country.sort_values(by='death_rate', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(death_rate_per_country['location'], death_rate_per_country['death_rate'], color='red')

plt.title("Top 10 Countries by COVID Death Rate")
plt.xlabel("Country")
plt.ylabel("Death Rate (Deaths per Case)")
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

# This cell shows total number of cases for the 20 countries with the most cases

```python
top_countries = df.groupby('location')['total_cases'].sum().reset_index()

top_countries = top_countries.sort_values(by='total_cases', ascending=False).head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_countries['location'], top_countries['total_cases'], color='skyblue')

plt.title("Top 20 Countries by Total COVID Cases")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# This cell shows  cumulative vaccinations over time for 20 countries

```python
top_countries = df.groupby('location')['total_cases'].sum().reset_index()
top_countries = top_countries.sort_values(by='total_cases', ascending=False).head(20)
top_country_list = top_countries['location'].tolist()

df_filtered = df[df['location'].isin(top_country_list)]

plt.figure(figsize=(12, 6))

for country in top_country_list:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title("Cumulative Vaccinations Over Time for Top 20 Countries")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

# This cell shows the Vaccination Rate for Top 20 Countries with Highest COVID Cases
# Yemen has the highest vaccination and india has the lowest

```python
top_countries = df.groupby('location')['total_cases'].sum().reset_index()
top_countries = top_countries.sort_values(by='total_cases', ascending=False).head(20)
top_country_list = top_countries['location'].tolist()

df_filtered = df[df['location'].isin(top_country_list)]

df_filtered['vaccination_rate'] = df_filtered['total_vaccinations'].div(df_filtered['population']).fillna(0) * 100

latest_vaccination_rate = df_filtered.groupby('location')['vaccination_rate'].max().reset_index()
latest_vaccination_rate = latest_vaccination_rate.sort_values(by='vaccination_rate', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(latest_vaccination_rate['location'], latest_vaccination_rate['vaccination_rate'], color='green')

plt.title("Vaccination Rate for Top 20 Countries with Highest COVID Cases")
plt.xlabel("Country")
plt.ylabel("Vaccination Rate (%)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.7)

plt.show()

```

# This cell shows the percentage vaccinated vs unvaccinated of the USA in a pie chart
# 66.7% of the population is vaccinated and 33.3% is unvaccinated

```python
df['vaccinated_population'] = df['total_vaccinations'].div(df['population']).fillna(0)
df['unvaccinated_population'] = 1 - df['vaccinated_population']

latest_vaccination = df.groupby('location')[['vaccinated_population', 'unvaccinated_population']].max().reset_index()


country = "United States" 
data = latest_vaccination[latest_vaccination['location'] == country]

if not data.empty:
    sizes = [data['vaccinated_population'].values[0], data['unvaccinated_population'].values[0]]
else:
    print(f"No vaccination data available for {country}")
    sizes = [0, 100] 

labels = ['Vaccinated', 'Unvaccinated']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title(f"Vaccination vs. Unvaccinated Population in {country}")

plt.show()
```
